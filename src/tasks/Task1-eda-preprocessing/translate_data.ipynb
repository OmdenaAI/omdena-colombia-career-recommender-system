## Brute force translation
I made a dictionart to translate the headers of shorter `saber_combined_all_fields.csv` and larger `saber_11_pro.csv` databases for this project, as well as the variables of the data (based on the shorter database).

### Translates values

```python
# open dictionary with dictionary of values for variables
file = open("../../data/translate_cat.txt", "r")
contents = file.read()
my_dics = ast.literal_eval(contents)
file.close()

# select cathegorical columns
cat_col=df.select_dtypes(include='object').columns.to_list()

# translates to English the values of the data frame
df_2 = df.copy()
for i in cat_col:
    try:
        df_2.replace({i: my_dics[i]}, inplace=True)
    except:
        next
```

### Translates header

```python
# open dictionary of translated headers
file = open("../../data/translate_headers.txt", "r")
contents = file.read()
file.close()

# get headers as dictionary 
my_headers = ast.literal_eval(contents) #capital letters
my_headers_low = ast.literal_eval(contents.lower()) #lower case

# translate header
df_2 = df_2.rename(columns=my_headers_low)#, inplace=True)
```
