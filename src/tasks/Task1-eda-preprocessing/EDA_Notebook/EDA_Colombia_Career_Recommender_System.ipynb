```python
null_df = (df.isna().sum() * 100 / len(df)).reset_index()
null_df.columns = ['Name', 'Unique_Count']
null_df.sort_values('Unique_Count', ascending=False).head(50)
```

```python
#drop column with more than 19%
drop_col_name = null_df[null_df["Unique_Count"]>19].Name.tolist()
df = df.drop(columns =drop_col_name)
```
