```python
# Read the text file
with open('translation_saber.txt') as f:
    lines = f.readlines()
```

```python
# Create a dictionary for the translations
translations = {}
for line in lines:
    key, value = line.split(",")
    key = key.strip("'")
    value = value.strip('\n')
    value = value[2:]
    translations[key] = value
```

```python
# Renaming the columns using the translations dictionary
df.rename(columns = translations, inplace = True)
```

```python
import itertools
 
def findsubsets(lst, n):
    return list(itertools.combinations(lst, n))
```

```python
saber_11 = ['Saber 11 ICFES Language', 'Saber 11 ICFES Mathematics ',
       'Saber 11 ICFES Biology ', 'Saber 11 ICFES Chemistry ',
       'Saber 11 ICFES Physics ', 'Saber 11 ICFES Social Sciences ',
       'Saber 11 ICFES Philosophy ', 'Saber 11 ICFES English']
```

```python
saber_11_subsets = findsubsets(saber_11, 2)
```

```python
cols = 2
rows = len(saber_11_subsets) // cols+1
fig, axs = plt.subplots(ncols=cols, nrows=rows, figsize=(20, 100))
plt.tight_layout(pad=3)
i = 0
for r in np.arange(0, rows, 1):
    for c in np.arange(0, cols, 1):
        if i >= len(saber_11_subsets):
            axs[r, c].set_visible(False)
        else:
            f1, f2 = saber_11_subsets[i]
            axs[r, c].scatter(df[f1], df[f2], alpha=0.5)
            t = f1 + " and " + f2
            axs[r, c].set_title(t, fontsize=15, pad=5)
        i+=1
```

```python
def display(feature):
    print("The Value counts for the feature: ")
    print(df[feature].value_counts())
    print("Are there any null values in the feature:", df[feature].isnull().any())
    if df[feature].isnull().any():
        null = df[df[feature].isnull()]
        print("The number of null values are:", len(null))
        print("The percentage of null values {:.2f}%".format((len(null)/len(df[feature]))*100))
        
def bar_plot_marks(feature):
    cols = 2
    rows = len(marks) // cols+1
    fig, axs = plt.subplots(ncols=cols, nrows=rows, figsize=(10,30))
    plt.tight_layout(pad=3)
    avg_lst = []
    std_lst = []
    u = df[feature].unique()
    uniq = [x for x in u if str(x) != 'nan']
    plt.legend(uniq)
    i = 0
    for r in np.arange(0, rows, 1):
        for c in np.arange(0, cols, 1):
            if i >= len(marks):
                axs[r, c].set_visible(False)
            else:
                for val in uniq:
                    data = df[df[feature] == val]
                    avg = data[marks[i]].mean()
                    std_lst.append(data[marks[i]].std())
                    avg_lst.append(int(avg))
                axs[r, c].bar(uniq, avg_lst, yerr=std_lst, alpha=0.5, ecolor='black', capsize=10)
                axs[r, c].set_title(marks[i], fontsize=15, pad=5)
                avg_lst = []
                std_lst = []
            i+=1
```
